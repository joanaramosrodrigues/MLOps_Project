### 2.2. Data Type Conversion


```python
# Convert host_since to datetime
listings['host_since'] = pd.to_datetime(listings['host_since'], errors='coerce')

# Convert price to numeric (if needed)
listings['price'] = pd.to_numeric(listings['price'], errors='coerce')

# Convert boolean values
bool_cols = ['host_is_superhost', 'host_has_profile_pic', 'host_identity_verified', 'instant_bookable']
for col in bool_cols:
    listings[col] = listings[col].map({'t': True, 'f': False})
```

### 2.3. Missing Values


```python
# Check missing value percentage
missing_percent = listings.isnull().mean().sort_values(ascending=False) * 100
print(missing_percent[missing_percent > 0])

```
